```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
from collections import Counter
```

```python
import nltk
nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
title_dataframe = pd.read_csv('./datasets/SparkAISummit2019/2019titlelist.csv', dtype={'title':'object'})
```

```python
title_dataframe['title'] = title_dataframe['title'].astype(str)
```

```python
word_list_dataframe = title_dataframe['title'].str.split(' ', expand=False)
```

```python
word_list = []

for label, content in word_list_dataframe.iteritems() :
    if hasattr(content, '__iter__') :
        for word in content :
            word_list.append(str(word).lower())
    else :
        word_list.append(str(content).lower())

word_list = [word for word in word_list if word not in stop_words]
bow_dataframe = pd.DataFrame(word_list, columns=['word'])
```

```python
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'white').generate(str(word_list))

fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
